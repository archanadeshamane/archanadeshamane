**Import liabraries**

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
from sklearn.pipeline import Pipeline
from sklearn.model_selection import GridSearchCV
from sklearn.preprocessing import StandardScaler
```

```python
house = pd.read_csv('/content/Housing.csv')
```

```python
house.head(5)
```

**One-hot encoding**

```python
house_enc = pd.get_dummies(house, drop_first=True)
```

```python
house_enc.head(5)
```

```python
x = house_enc.drop('price',axis=1)
y = house_enc['price']
```

**Train-Test split**

```python
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.2,random_state=40)
```

**Standardization**

```python
scaler= StandardScaler()
```

```python
x_trn_scl = scaler.fit_transform(x_train)
x_tst_scl = scaler.transform(x_test)
```

**Training the model**

```python
lr=LinearRegression()
```

```python
poly=PolynomialFeatures()
```

```python
pipe = Pipeline([('POLY',poly),('LR',lr)])
```

```python
pipe.get_params()
```

```python
params = {'POLY__degree':[1,2,3,4]}
model= GridSearchCV(pipe, param_grid=params)
```

```python
model.fit(x_trn_scl,y_train)
```

**Model evaluation**

```python
print("Best Score:", model.best_score_)
print("Best degree=", model.best_params_)
print("Best degree =", model.best_params_['POLY__degree'])
```

**Model prediction**

```python
y_pred= model.predict(x_tst_scl)
```
